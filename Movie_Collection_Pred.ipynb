```python
# Outliers treatment for Marketing Expense
```

```python
np.percentile(df.Marketin_expense,[99])[0]
```

```python
uv=np.percentile(df.Marketin_expense,[99])[0]
```

```python
df.Marketin_expense[(df.Marketin_expense > 3*uv)]=3*uv
```

```python
# Outliers treatment for twitter hashtags
```

```python
np.percentile(df.Twitter_hastags,[99])[0]
```

```python
lv=np.percentile(df.Twitter_hastags,[99])[0]
```

```python
df.Twitter_hastags[df.Twitter_hastags > lv]
```

```python
df.Twitter_hastags[(df.Twitter_hastags > 3*lv)]=3*lv
```

```python
df.describe()
```

```python
# Outliers treatment for Average age actors
```

```python
np.percentile(df.Avg_age_actors, [1])[0]
```

```python
mv=np.percentile(df.Avg_age_actors,[1])[0]
```

```python
df.Avg_age_actors[(df.Avg_age_actors < mv)]
```

```python
df.Avg_age_actors[(df.Avg_age_actors < 0.3*mv)]=0.3*mv
```

```python
df.describe()
```

```python
# Missing Values treatment for Time Taken
```

```python
df.info()
```

```python
df.Time_taken=df.Time_taken.fillna(df.Time_taken.mean())
```

```python
X = df.drop('Collection',axis=1)
y = df['Collection']
```

```python
X.shape,y.shape
```

```python
 from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree = DecisionTreeRegressor(random_state=17)
```

```python
tree.fit(X_train, y_train)
```

```python
tree.score(X_train,y_train)
```

```python
tree.score(X_test,y_test)
```

```python
y_pred = tree.predict(X_test)
y_pred
```
